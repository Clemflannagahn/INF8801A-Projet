```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# print(matplotlib.rcParams.keys())
matplotlib.rcParams.update({'font.size': 16})
matplotlib.rcParams.update({'font.sans-serif': ['Noto Sans']})
d1 = [27, 35, 43, 49, 51, 54, 55, 55, 59, 59]
d2 = [65, 71, 73, 78, 78, 81, 85, 87, 87, 89]
d3 = [60, 64, 69, 72, 74, 77, 77, 78, 80, 83]
X = np.arange(1,11)
plt.plot(X, d1, label='d1')
plt.plot(X, d2, label='d2')
plt.plot(X, d3, label='d3')
plt.legend()
plt.xticks(X)
plt.xlabel('k')
plt.ylabel('Part de prsence dans le top k')
plt.title('Part de prsence dans le top k des 3 distances')
plt.show()
```
