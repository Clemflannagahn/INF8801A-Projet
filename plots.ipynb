```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
IMG_DIR = 'images'
os.makedirs(IMG_DIR, exist_ok=True)
matplotlib.rcParams.update(
    {
        'font.size': 16,
        'font.family': 'serif',
        'font.serif': ['Times New Roman'],
        'text.usetex': True,
        'legend.framealpha' : 0.7,
        'figure.facecolor': '#37474F',
        'axes.facecolor': '#37474F',
        'axes.edgecolor': '#D6D6D6',
        'axes.labelcolor': 'black',
        'legend.facecolor': '#37474F',
        'text.color': '#D6D6D6',
        'xtick.color': '#D6D6D6',
        'ytick.color': '#D6D6D6',
        'axes.labelcolor': '#D6D6D6',
    }
)
```

## Distances implémentées

```python
d1 = [27, 35, 43, 49, 51, 54, 55, 55, 59, 59]
d2 = [65, 71, 73, 78, 78, 81, 85, 87, 87, 89]
d3 = [60, 64, 69, 72, 74, 77, 77, 78, 80, 83]

X = np.arange(1,11)

plt.plot(X, d1, label=r'$d_1$', color='#9EC3E6', linestyle='--')
plt.plot(X, d2, label=r'$d_2$', color='#B6D7A8')
plt.plot(X, d3, label=r'$d_3$', color='#F9CB9C', linestyle='-.')

plt.xlim(1, 10)
plt.ylim(0, 100)
plt.legend().set_title('Distance')
plt.xticks(X)

plt.xlabel(r'$k$')
plt.ylabel(r'Part de présence dans le top $k$')
plt.title(r'Images dans le top $k$ (Implémentatation)')

plt.savefig(os.path.join(IMG_DIR,'base_impl.png'), dpi=600)
plt.show()
```
